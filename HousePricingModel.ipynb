```python
import pandas as pd
DF_train=pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
DF_train
```

```python
DF_train_final=DF_train.loc[:,['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']]
DF_train_final
```

```python
x_train=DF_train_final.loc[:,['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
x_train
```

```python
y_train=DF_train_final.loc[:,['median_house_value']]
y_train
```

```python
from sklearn.linear_model import LinearRegression
model_LR=LinearRegression()
model_LR.fit(x_train, y_train)
```

```python
DF_test=pd.read_csv('/content/sample_data/california_housing_test.csv')
DF_test
```

```python
x_test=DF_test.loc[:,['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
x_test
```

```python
y_test=DF_test.loc[:,['median_house_value']]
y_test
```

```python
from sklearn.linear_model import LinearRegression
model_LR= LinearRegression()
model_LR.fit(x_test, y_test)
```

```python
linear_predict=model_LR.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
linear_mae = metrics.mean_squared_error(y_test, linear_predict)
print('Mean Square Error:', metrics.mean_squared_error(y_test, linear_predict))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, linear_predict))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, linear_predict)))
```

```python
from sklearn.svm import SVR
model_svr = SVR(kernel = 'linear')
model_svr.fit(x_train, y_train)
```

```python
svr_predict=model_svr.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
svr_mae = metrics.mean_squared_error(y_test, svr_predict)
print('Mean Square Error:', metrics.mean_squared_error(y_test, svr_predict))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, svr_predict))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, svr_predict)))
```

```python
from sklearn.svm import SVR
model_rbf = SVR(kernel = 'rbf')
model_rbf.fit(x_train, y_train)
```

```python
rbf_predict=model_rbf.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
rbf_mae = metrics.mean_squared_error(y_test, rbf_predict)
print('Mean Square Error:', metrics.mean_squared_error(y_test, rbf_predict))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, rbf_predict))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, rbf_predict)))
```

```python
from sklearn.svm import SVR
model_poly = SVR(kernel = 'poly')
model_poly.fit(x_train, y_train)
```

```python
poly_predict=model_poly.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
poly_mae = metrics.mean_squared_error(y_test, poly_predict)
print('Mean Square Error:', metrics.mean_squared_error(y_test, poly_predict))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, poly_predict))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, poly_predict)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
x_axis = ['model_LR', 'model_svr', 'model_rbf', 'model_poly']
y_axis = [linear_mae, svr_mae, rbf_mae, poly_mae]

plt.bar(x_axis, y_axis)
plt.xlabel("Models")
plt.ylabel("Mean Squared Error")
plt.title("Models for California Housing")

# plt.grid(linestyle='--')
plt.xticks(fontsize=7)

plt.show()
```
